```
import numpy as np
import pandas as pd
import textblob
from textblob import TextBlob
import re
```

```
data = pd.DataFrame(data=[tweet.text for tweet in tweets], columns=['Tweets'])
display(data.head(10))
```

```

def clean_tweet(tweet):
  return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|      (\w+:\/\/\S+)"
                         , " ", tweet).split())


```

```
def analize_sentiment(tweet):
  analysis = TextBlob(clean_tweet(tweet))
  if analysis.sentiment.polarity > 0:
      return 1
  elif analysis.sentiment.polarity == 0:
       return 0
  else:
       return -1
    
    
    
    
    
    
    
    
    
```

```


data['SA'] = np.array([ analize_sentiment(tweet)
                       for tweet in data['Tweets'] ])

```

```
pos_tweets = [ tweet for index, tweet in enumerate(data['Tweets']) 
              if data['SA'][index] > 0]
neu_tweets = [ tweet for index, tweet in enumerate(data['Tweets']) 
              if data['SA'][index] == 0]
neg_tweets = [ tweet for index, tweet in enumerate(data['Tweets']) 
              if data['SA'][index] < 0]

# Printando as porcentagens:

print("Porcentagem de tweets positivos: {}%".
      format(len(pos_tweets)*100/len(data['Tweets'])))
print("Porcentagem de tweets neutros: {}%".
      format(len(neu_tweets)*100/len(data['Tweets'])))
print("Porcentagem de tweets negativos: {}%".
      format(len(neg_tweets)*100/len(data['Tweets'])))



```
